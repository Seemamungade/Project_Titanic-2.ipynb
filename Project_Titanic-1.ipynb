```python
# Pclass 
titanic_df["Pclass"].value_counts()
```

```python
#ClassSurvival
class_sur = pd.crosstab(titanic_df['Pclass'],titanic_df['Survived'])
class_sur
```

```python
class1Sur = (class_sur.loc[1][1]*100 )/(class_sur.loc[1].sum())
class1Sur
```

```python
class2Sur = (class_sur.loc[2][1]*100)/(class_sur.loc[2].sum())
class2Sur
```

```python
class3Sur = (class_sur.loc[3][1]*100)/(class_sur.loc[3].sum())
class3Sur
```
